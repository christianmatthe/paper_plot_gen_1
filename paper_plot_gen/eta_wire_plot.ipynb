### Normalize to average of etas (average weight of 1)

```python
z_space = np.linspace(-10,10,num=900)


eta_space = np.array([eta_wire(z) if (z < 10 and z > -10) else 0 for z in z_space])
eta_space = eta_space/np.average(eta_space)


fig = plt.figure(0, figsize=(8,6.5))
ax1=plt.gca()
ax1.plot(z_space, eta_space,color = "k" ,label = r"$\eta_{wire}$",
         lw= 10)


ax1.set_xlabel(r"x [mm]")
ax1.set_ylabel(r"A.U.")

plt.legend(shadow=True)
plt.tight_layout()
plt.grid(True)
```
